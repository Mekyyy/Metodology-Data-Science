##Dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
print("Banyaknya data: ", len(X))
```

```python
print(X[:10]) #buat ngambil data bentuk list
```

```python
print(y[:10])
```

##Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8)

print("Banyak data latih setelah dilakukan Train-Test : ", len(X_train))
print("Banyak data uji setelah dilakukan Train-Test : ", len(X_test))
```

##K_FOLD VALIDATION
Hanya untuk datasets sedikit

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

model = SVC(kernel = 'linear', C = 1)
scores = cross_val_score(model, X, y, cv = 5)
print("Akurasi model SVM untuk tiap fold : ", scores)
print("Akurasi model SVM dengan 5-fold Cross Validation: ", scores.mean())
```

#MODEL

#1. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan KNN : ", score)
```

#2. D-TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

dt = DecisionTreeClassifier(max_depth=None, min_samples_split=2)

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan D-TREE : ", score)
```

Hyper Parameter Tunning

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

dt = DecisionTreeClassifier(max_depth=3, min_samples_split=2)

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan D-TREE : ", score)
```

#3. Naive Bayes

```python
from sklearn import naive_bayes
from sklearn import  metrics

nb = naive_bayes.BernoulliNB()

nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan menggunakan Naive Bayes: ", score)
```

#4. SVM

```python
#SVM RBF
from sklearn.svm import SVC
from sklearn import metrics

svm = SVC(kernel = 'rbf', C = 1, gamma = 0.01)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan menggunakan SVM RBF: ", score)
```

```python
#SVM LINEAR
from sklearn.svm import SVC
from sklearn import metrics

svm = SVC(kernel = 'linear', C = 1)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan menggunakan SVM RBF: ", score)
```

```python
#SVM POLY
from sklearn.svm import SVC
from sklearn import metrics

svm = SVC(kernel = 'poly', C = 1, gamma = 0.01, degree = 2)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi dengan menggunakan SVM RBF: ", score)
```

#5. ADAPTIVE BOOSTING

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics

ab = AdaBoostClassifier(n_estimators=50, learning_rate = 1)

ab.fit(X_train, y_train)
y_pred  = ab.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
print(score)
```
