#2. Bar Chart

```python
import matplotlib.pyplot as plt


nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]


total_permen = sum(jumlah_permen)


jumlah_kopiko = jumlah_permen[nama_permen.index('Kopiko')]


persentase_kopiko = (jumlah_kopiko / total_permen) * 100


plt.figure(figsize=(8, 6))
plt.bar(nama_permen, jumlah_permen, color=['blue', 'orange', 'green', 'red', 'purple'])
plt.xlabel('Jenis Permen')
plt.ylabel('Jumlah Permen')
plt.title('Jumlah Permen Berdasarkan Jenis')
plt.xticks(rotation=45)

plt.text(1, jumlah_kopiko + 5, '{:.2f}%'.format(persentase_kopiko), ha='center')

plt.show()

print("Peluang memilih permen Kopiko: {:.2f}%".format(persentase_kopiko))
```
